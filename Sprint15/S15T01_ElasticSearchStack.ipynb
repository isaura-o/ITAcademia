## Data

The dataset choosen for this sprint ilustrates how change the temperatures because of the [Climate Change](https://www.kaggle.com/datasets/sevgisarac/temperature-change).

The dataset it is comprised of 66 features. 

- Area: in which the temperature was taken and numerical value associated. 
- Months: in which the value was taken (January to December) or trimester (i.e., December-January-February) or if the temperature difference it is yearly taken. 
- Unit: feature which indicates that the difference of temperature was taken in Centigrads. 
- The rest of columns are the observations of difference of temperature of the years from 1961 to 2019.
