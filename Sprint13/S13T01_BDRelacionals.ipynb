The following figure shows a the diagram of relational base that contains: Clients, Orders (compres), Employees (workers), Establishment, products, and, products types.

For this exercise we have choose, first, to show the diagram entity-relation, and then, since they are asking for the information on each table, we will use the typical "table diagram", like the one at the end of this [page](https://www.chrisbell.com/SNHU/IT-510-Advanced-Information-Technology/database-management-system-DBMS-relationships.php).

The **squares** in the diagram are the **Entities**, the **circles** are the **Attributes**. This is why in this case, since product type is a table which contains the description of the products, we use the circle for product type and the square for product.

The **diamond** shape is reserved for the attributes that **interrelate** two or more entities. In this case we have two tables that are able to interrelate: Order and Establishment. Also, we need to noticies that not only they can iterrelate entities also,they are permitted to interrelate with another interrelational table.

Looking at the figure above, we have concluded that client, products, and, employees are by themselves entities. Because each one of these contains a number of attributes like we see in the case of product that have the attribute product type, which is an attribute multivaluated, since may contain the product name, brand, description, etc. (we will see this in the next part). The five entities: client, product, employee, order, and, establishment. Product and order are related between themselves with the interrelational relationship "has", this is because order and product are a relationship many-to-many, the other relationships we can see that they are one-to-many, and they don't need an specific interrelated entity. Client and Order are related with the interrelation "performs", Employee and Order with "makes", and, Product and Establishment with "are", since the products are on the establishments.

Now, let's characterize each table with its attributes, we also will draw their property relations: 1:1, 1:n, and, n:n. To do, so instead of using the diagram like before, it is more confortable use the Relational Diagram, but we will be using Chen's notation from the ER diagrams, like the one it follows, because of the large number of attributes.

In this figure, we can see that inside that each entity/attribute we have write some attributes that belong to their parent table, and also that may belong to a another entity/attribute if the attribute it is multivaluated. Single head arrows mean that the relation is 1:1, while double headed arrows mean that the relation 1:n. For n:n we also have used double headed double arrows.
Also, we have write down the relationships of the tables.

- The relationship between clients and orders it is 1:n, since a client may place multiple orders.
  We have choose for clients the attributes: Name, Last Name, Address 1, Address 2 (if more addresses are needed), and, phone number.

- The products to orders it is n:n, since multiple orders can have multiple products, and they are related with the interrelational relationship "has".
  We have choose for products the attributes: Item_ID, Establishment_ID, Unit price.

- The establishments-employees it is 1:1, the first, since each employee it is on one establishment, while it is 1:n for establishment-product, since a product may be in more than one establishment.
  We have choose for establishment: Name, Address, Phone number, and, surface. For Employees: Name, Last Name, Title, and, work phone number. And for Order: Client_ID, Employee_ID, Establishment_ID, Product_ID, Quantity, and, Total Price.

- The products-product type is a 1:1, since each item has specific description, brand, etc. 
  For product type: Name Product, brand, and, description of the item.
